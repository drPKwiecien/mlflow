```python
import mlflow

import pandas as pd
from mlflow import MlflowClient
client = MlflowClient(tracking_uri="http://127.0.0.1:5000")

all_experiments = client.search_experiments()

print(all_experiments)

```

```python
summary_data = []

for experiment in all_experiments:
    runs = client.search_runs(experiment_ids=[experiment.experiment_id], filter_string="")
    
    for run in runs:
        experiment_name = experiment.name
        run_name = run.data.tags.get('mlflow.runName', "N/A")  # Adjust based on how you've named your runs
        accuracy_score = run.data.metrics.get('accuracy', None)  # Adjust if using a different metric name
        
        summary_data.append({
            "Experiment Name": experiment_name,
            "Run Name": run_name,
            "Accuracy Score": accuracy_score
        })

summary_df = pd.DataFrame(summary_data)

summary_df.sort_values(by="Accuracy Score", ascending=False)
```
